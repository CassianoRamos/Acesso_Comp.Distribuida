```python
df = pd.read_csv('logs.csv', encoding='utf-8')
df.head()
```

```python
#Quantidade de acessos
entradas_autorizadas = df[(df[' Permissão'] == 'Usuário autorizado') & (df[' Entrada/Saída'] == 'Entrada')]


usuarios_unicos = entradas_autorizadas[' Número da tag'].nunique()

print(f"Quantidade de pessoas que acessaram a sala: {usuarios_unicos}")
```

```python
#Tempo de permanencia na sala
entradas = df[df[' Entrada/Saída'] == 'Entrada'].copy()
saidas = df[df[' Entrada/Saída'] == 'Saída'].copy()

entradas.reset_index(drop=True, inplace=True)
saidas.reset_index(drop=True, inplace=True)

tempo_permanencia = saidas['Data/Hora'] - entradas['Data/Hora']

for idx, tempo in enumerate(tempo_permanencia):
    rfid = entradas.iloc[idx][' Número da tag']
    nome = entradas.iloc[idx][' Nome do Usuário']
    print(f"Colaborador: {nome} (RFID: {rfid}) ficou na sala por {tempo}")
```

```python
#Tentativas de invasão
tentativas_invasao = df[df[' Nome do Usuário'].str.contains('Tentativas de invasão', na=False)]

total_tentativas = tentativas_invasao.shape[0]

print(f"Quantidade total de tentativas de invasão: {total_tentativas}")
```

```python
#Tentativas de acesso não autorizado
tentativas_nao_autorizada = df[df[' Permissão'].str.contains('Usuário não autorizado', na=False)]

total = tentativas_nao_autorizada.shape[0]

print(f"Quantidade total de tentativas de acesso não autorizado: {total}")
```
